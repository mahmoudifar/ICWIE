```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Wedge
import matplotlib
matplotlib.rcParams['figure.dpi']=300


def Plot_Vline(ax , YMIN, YMAX, X , color, linestyle , lw =3, alpha = 1 ): 
    ax.vlines(X, ymin=YMIN, ymax=YMAX, colors=color\
              ,zorder = 21, linestyle=linestyle , lw =lw, alpha = alpha)  
    
def Plot_Hline(ax , XMIN, XMAX, Y , color, linestyle,alpha=1, lw = 3): 
    ax.hlines(Y, xmin=XMIN, xmax=XMAX, colors=color, zorder = 22, lw =lw, linestyle=linestyle, alpha=alpha)

fig, ax =plt.subplots(ncols=1, nrows=1, figsize=(10,10), facecolor='w')
r = 0.35  
n = 5  
angles = np.linspace(0, 2 * np.pi, n+1, endpoint=True)
x = r * np.cos(angles)
y = r * np.sin(angles)
color = ['w', 'k', 'w', 'k', 'w']
cv = ['k', 'w', 'k', 'w', 'k']
text = ["$v_1$", "$v_2$", "$v_3$", "$v_4$","$v_5$" ]
for i in range(n):
    ax.plot(x[i], y[i])
    ax.text(x[i], y[i], text[i], color = cv[i], ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')
    circle = plt.Circle((x[i], y[i]), 0.05, fc = color[i], zorder = 4, ec = 'k')
    ax.add_patch(circle)
ax.plot(x, y, color = 'k')

lw = 0.1
y = [0.2, 0, -0.2, -0.7, -0.9 , -1.1, -0.7, -0.9, -1.1, 0.2, 0, -0.2, -0.7, -0.9, -1.1]
alpha = [1 , 0.3 , 0.3, 1 , 0.3, 0.3, 1, 0.3, 0.3, 1 , 0.3 , 0.3, 1 ,0.3 , 0.3]
ls = ['-' , '--' , '--', '-' , '--', '--', '-', '--','--', '-' , '--' ,'--', '-' ,'--' ,'--']

x1 = [0.5]*6+[-0.3]*3 +[1.5]*6
x2 = [0.6]*6+[-0.2]*3 +[1.6]*6

xr = [0.6, 1.6, -0.2, 0.6 , 1.6]
yr = [-0.25, -0.25, -1.15, -1.15, -1.15]
z1 = [0, 0, -0.9, -0.9 , -0.9]*5
for i in range(len(y)):
    Plot_Hline(ax , x1[i], x2[i], y[i] , '#2E8B57', linestyle =ls[i], alpha=alpha[i], lw =1)
    
for i in range(len(xr)):
    ax.add_patch(Rectangle(( xr[i], yr[i]), 0.15, 0.5,fc = '#CAE1FF', \
                           ec = 'k', angle = 0, zorder =25))
    ax.text(xr[i]+0.15/2, z1[i], '$U_i$', color = 'k', ha ='center',
                                     va ='center', zorder = 26, fontsize = 12, math_fontfamily='dejavuserif')
    
x3 = [0.75]*6 +[-0.05]*3+[1.75]*6 
x4 = [0.85]*6+[0.05]*3+[1.85]*6

for i in range(len(y)):
    Plot_Hline(ax , x3[i], x4[i], y[i] , '#2E8B57', linestyle ='-', alpha=1, lw =1)
    ax.add_patch(Rectangle(( x4[i], y[i]-0.05/2), 0.1, 0.05,fc = '#AEEEEE', \
                           ec = 'k', angle = 0, zorder =25))
    
x5 = [0.95]*6+[0.15]*3+[1.95]*6
x6 = [1.05]*6+[0.25]*3+[2.05]*6
alpha1 = [1 , 1 , 1, 0.3 , 1, 1,  0.3, 0.3, 1, 0.3 , 0.3 , 1, 1 ,1 , 1]
for i in range(len(y)):
    Plot_Hline(ax , x5[i], x6[i], y[i] , "#CD2626", linestyle ='-', alpha=alpha1[i], lw =1)

xpath = [0.8, 1.8, 0, 0.8, 1.8]
ypath = [-0.05, -0.05, -1.15 , -1.15, -1.15]

for i in range(len(xpath)):
       ax.add_patch(Rectangle(( xpath[i], ypath[i]), 0.2, 0.3,fc =  'lightgray', \
                           ec = 'none', angle = 0, zorder =5))
    
y1 = [0.2]*2+[-0.7]*3
y2 = [-0.125]*2+[-0.825]*3

x = [0.9, 1.9, 0.1, 0.9 , 1.9]
alpha2 = [1, 1, 1, 1 ,0.3]
dc = ['w', 'k', 'w', 'k', 'w']
namefd = [r"$D_{i_{1,2}}}$", r"$D_{i_{1,2}}}$", r"$D_{i_{1}}}$", r"$D_{i_1}$", 'rf']
for i in range(len(y1)):
    Plot_Vline(ax , y1[i], y2[i], x[i] , '#1874CD', linestyle = '-' , lw =1, alpha =alpha2[i] )
    if alpha2[i]>0.3:
        ax.add_patch(Wedge(( x[i], y2[i]+0.025 ), 0.03,180, 0,fc = dc[i], ec = 'k', zorder = 28,lw =1))
        Plot_Vline(ax , y2[i], y2[i]-0.04, x[i] , 'k', linestyle = '-' , lw =1, alpha =1 )
        ax.text(x[i]+0.09, y2[i], namefd[i], color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')

y3 = [-0.2, -0.2, -0.9, -0.9, -0.9]
y4 = [-0.325, -0.325 , -1.225, -1.225, -1.225]
alpha3 = [0.3, 0.3, 0.3 , 1 , 1]
dc = ['w', 'k', 'w', 'w', 'k']
named = ['x', 'x', 'x', r"$D_{i_{2,3}}}$", r"$D_{i_{2,3}}}$" ]

for i in range(len(y3)):
    Plot_Vline(ax , y3[i], y4[i], x[i] , '#1874CD', linestyle = '-' , lw =1, alpha = alpha3[i] )
    if alpha3[i]>0.3:
        ax.add_patch(Wedge(( x[i], y4[i]+0.025 ), 0.03,180, 0,fc = dc[i], ec = 'k', zorder = 28,lw =1))
        Plot_Vline(ax , y4[i], y4[i]-0.04, x[i] , 'k', linestyle = '-' , lw =1, alpha =1 )
        ax.text(x[i]+0.09, y4[i], named[i], color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')
    
xu = [1.05, 1.05, 2.05]
yu = [-0.05, -1.15, -1.15]
lu = [0.3,0.3, 0.5]
d1 = [0.1, -1, -0.9]
colu = ['#EAEAEA', '#E0FFFF', '#B0C4DE']

for i in range(len(xu)):
    ax.add_patch(Rectangle((xu[i], yu[i]), 0.15, lu[i],fc = colu[i], \
                           ec = 'k', angle = 0, zorder =25))
    ax.text(xu[i]+0.15/2, d1[i], r"$U_f^{\dagger}$", color = 'k', ha ='center',
                                     va ='center', zorder = 25, fontsize = 12, math_fontfamily='dejavuserif')

    
yf = [0.2, 0 , -0.9 , -1.1, -0.7, -0.9 , -1.1]   
xf1= [1.2, 1.2, 1.2, 1.2, 2.2, 2.2 , 2.2 ]
xf2 = [1.3]*4+[2.3]*3
alpha3 = [0.3 , 0.3, 0.3,0.3,1, 0.3, 0.3]
col = ['k', 'k', 'k', 'k', 'w', 'k', 'k']
nnd = ['k', 'k', 'k', 'k', r'$D_{post}$', 'k', 'k']

for i in range(len(yf)):
    Plot_Hline(ax , xf1[i], xf2[i], yf[i] , "#CD2626", linestyle ='-', alpha= alpha3[i], lw =1)
    if alpha3[i]>0.3:
        ax.add_patch(Wedge(( xf2[i], yf[i] ), 0.03,-90, 90,fc = col[i], ec = 'k', zorder = 28,lw =1))
        Plot_Hline(ax , xf2[i], xf2[i]+0.065, yf[i] , "k", linestyle ='-', alpha= 1, lw =1)
        ax.text(xf2[i]+0.03,  yf[i]-0.07, nnd[i], color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')    
        
yo = [0.1, 0.1 , -1 , -1 , -1, -0.815 ]
xo = [0.9, 1.9, 0.1, 0.9 , 1.9,1.9]
for i in range(len(xo)):
    ax.add_patch(Rectangle((xo[i]-0.025, yo[i]-0.015), 0.05, 0.015,fc = 'k', \
                           ec = 'k', angle = 0, zorder =25))
        
yd = [-0.2, -0.2, -1.1]
xd = [1.05, 2.05, 0.25]
col1 = ['k', 'w', 'k']
nd = [r'$D_{post}$', r'$D_{s_3}$', r'$D_{s_3}$']
for i in range(len(yd)):
    ax.add_patch(Wedge(( xd[i], yd[i] ), 0.03,-90, 90,fc = col1[i], ec = 'k', zorder = 28,lw =1))
    Plot_Hline(ax , xd[i], xd[i]+0.065, yd [i],'k' , linestyle ='-', alpha= 1, lw =1)
    ax.text(xd[i]+0.03,  yd[i]-0.07, nd[i], color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')
    
ax.add_patch(Rectangle((0.4, -1.6), 1.3, 0.3,fc = 'none', \
                           ec = 'k', angle = 0, zorder =25, linestyle ='--'))
ax.add_patch(Rectangle((0.55, -1.52), 0.1, 0.05,fc = '#AEEEEE', \
                           ec = 'k', angle = 0, zorder =25))

ax.text(0.6, -1.38, 'Nonlinear\n Crystal ', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')
ax.add_patch(Wedge(( 0.9, -1.48 ), 0.04,90, -90,fc = 'w', ec = 'k', zorder = 28,lw =1))
Plot_Hline(ax , 0.9, 0.9-0.075, -1.48 , "k", linestyle ='-', alpha= 1, lw =1)
ax.add_patch(Wedge(( 1, -1.48 ), 0.04,-90, +90,fc = 'k', ec = 'k', zorder = 28,lw =1))
Plot_Hline(ax , 1, 1+0.075, -1.48 , "k", linestyle ='-', alpha= 1, lw =1)

ax.text(0.95, -1.38, 'Detector', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')

ax.text(0.88, -1.56, 'on', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 10, math_fontfamily='dejavuserif')
ax.text(1.02, -1.56, 'off', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 10, math_fontfamily='dejavuserif')

ax.add_patch(Rectangle((1.2, -1.58), 0.1, 0.15,fc = 'lightgray', \
                           ec = 'none', angle = 0, zorder =25))
ax.text(1.25, -1.37, 'Path\n Identity ', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')

ax.add_patch(Rectangle((1.5, -1.52), 0.1, 0.03,fc = 'k', \
                           ec = 'k', angle = 0, zorder =25))
ax.text(1.55, -1.38, 'Object', color = 'k', ha ='center',
                                     va ='center', zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')

h1 = [0.48, 0.48, 1.48, 1.48, -0.32]
h2 = [1.32, 1.32, 2.12, 2.38, 0.32]
v1 = [0.3, -0.6, 0.3, -0.6, -0.6]

pro = [r'$P(0,1|1,2)$',r'$P(0,1|4,5)$', r'$P(0,1|2,3)$', r'$P(0,1|5,1)$',r'$P(0,1|3,4)$'  ]

for i in range(len(h1)):
    Plot_Hline(ax , h1[i], h2[i], v1[i] , "k", linestyle ='-', alpha= 1, lw =1)
    ax.text( h1[i]+0.02, v1[i]+0.025, pro[i], color = 'k',  zorder = 5, fontsize = 12, math_fontfamily='dejavuserif')
ax.text( 1.93 ,-0.835, r'$t=0$', color = 'k',  zorder = 5, fontsize = 10, math_fontfamily='dejavuserif')
ax.set_title('(a)',x=0.03, y=1.03, fontsize = 12)
ax.set_aspect(1)
ax.set_xlim([-0.35, 2.4])
ax.set_ylim([-1.65, 0.42])
ax.axis('off')
plt.show()   
fig.savefig("pro.png", bbox_inches='tight')
```
